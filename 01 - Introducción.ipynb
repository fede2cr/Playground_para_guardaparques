# Introducción

Vamos a aprender como programar con las CircuitPlayground de Adafruit

## Conociendo la Playground

![Playground](https://raw.githubusercontent.com/fede2cr/Playground_para_guardaparques/master/img/circuit_playground_express-labeled.jpg)

Las playground son una excelente opción para aprender sobre electrónica, robótica y programación dado que es una placa con múltiples sensores integrados de forma que no debemos preocuparnos por conectar cables o por quemar dispositivos como sucede con placas como las Arduino UNO.

Un excelente recurso (en inglés) es la [guía de Adafruit](https://learn.adafruit.com/adafruit-circuit-playground-express).

Podemos observar:

- Sensor de luz
- Sensor de movimiento
- Sensor de temperatura
- Micrófono
- Switch
- Dos botones
- 10 luces multi-color (neopixel)
- 1 luz normal (en pin D13)
- Emisor infrarojo
- Receptor infrarojo
- 14 pines para conectar sensores y actuadores
- Conector USB
- Conector para baterías

## Programando la Playground

La forma más sencilla, clara y rápida de programar una Playground es por medio del sitio de [makecode.adafruit.com](https://makecode.adafruit.com/) donde vamos a trabajar las acciones que deseamos de nuestra tarjeta, por medio de lenguaje de bloques orientado a **eventos**.

Podemos observar algunas de las categorías de bloques que tenemos disponibles:

![Categorías de bloques](https://raw.githubusercontent.com/fede2cr/Playground_para_guardaparques/master/img/Input.png)

En este caso podemos ir a la categoría de **Light** y conectar al bloque de **Forever** una acción para activar las luces multicolor de la placa.

### Laboratorio "Hola mundo"

1. Saque de la categoría Light, un bloque de Show Ring

![Show Ring, sin conectar](https://raw.githubusercontent.com/fede2cr/Playground_para_guardaparques/master/img/show-ring-sin-conectar.png)

2. Conecte al bloque Forever

3. Observe como el emulador de Playground presenta las acciones al lado izquierdo.

4. Para programar nuestra Playground, damos click en **Download** y luego Aceptar a la hora de Descargar el archivo.

5. Conectamos la Playground, y presionamos el botón de reset **dos veces**, donde prende los LEDs de color Verde, y aparece un disco con nombre **CPLAYBOOT**. Nota: Si las luces prenden de color rojo y el disco no aparece, revise que el cable USB sea para datos y carga, y no solo para carga.

6. Arrastre el archivo descargado hacia el disco CPLAYBOOT y observe como la Playground carga el código y se comporta de forma idéntica al emulador de la página web.

![Playground Hello World](https://raw.githubusercontent.com/fede2cr/Playground_para_guardaparques/master/img/Playground%20hello%20world.png)

**Opcional**: Cambie **Show Ring** por **Show Animation** y juegue un poco con las opciones disponibles.

## Eventos por entradas

En la categoría de **Input** es donde podemos crear los eventos que controlan el comportamiento de la placa. Podemos observar:

- Cuando botón
- Cuando se bata (movimiento)
- Cuando un sonido duro
- Cuando la luz esté oscura/brillante
- Cuando el switch esté izquierda/derecha

![Categoría de Input](https://raw.githubusercontent.com/fede2cr/Playground_para_guardaparques/master/img/Input%20blocks.png)

### Laboratorio "Luz de noche"

Utilizando los eventos por entradas, cree un código que funcione como una luz de noche:

1. Cuando el sensor de luz diga que está obscuro, debe prender el anillo de luces o mostrar una animación con las luces.
2. Cuando el sensor de luz diga que está claro, debe apagar las luces.

Para realizar esto, creamos dos bloques de tipo Input, cada uno con la condición de luz correcta, y dentro un bloque de Show Ring ya sea con las luces de color, o con las luces de color "negro" para apagarlas, así como también puede usar un bloque llamado **Clear**.
