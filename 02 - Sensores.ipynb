# Sensores

Ahora vamos a aprender como conectar sensores a un Playground.

## Digital Read y Write

Cuando manipulamos pines de forma individual, por ejemplo para prender una luz de tipo LED, utilizamos funciones como DigitalRead y DigitalWrite.

![Digital Read y Write](https://raw.githubusercontent.com/fede2cr/Playground_para_guardaparques/master/img/Pins.png)

### Laboratorio de LED

Vamos a prender el LED pequeño de color rojo que se encuentra al lado derecho del puerto USB.

1. Vamos a utilizar el bloque de **Forever** pero se recomienda experimentar con bloques de eventos.
2. A este bloque le agregamos un DigitalWrite, el cual configuramos para que use el pin de nombre **"LED"**. Para prender y apagar el LED, vamos a utilizar dos bloques, uno para prender el LED o **"HIGH"** y para apagarlo en **"LOW"**.
3. Las microcontroladoras trabajan a velocidades muy rápidas para el humano... Si queremos ver cuando prender y apagar la luz, debemos utilizar bloques de **"Pause"** en medio de cada DigitalWrite. Se recomienda jugar con el valor pausa, y observar el comportamiento del LED.

![Laboratorio de LED](https://raw.githubusercontent.com/fede2cr/Playground_para_guardaparques/master/img/Pin%20D13.png)

## Conectando sensores

Para conectar sensores podemos utilizar:

- Soldadura para conexiones permanentes
- Hilo conductivo, cinta conductiva para electrovestibles
- Tornillos para sujetar conexiones temporales o semipermanentes

Existen algunos sensores que utilizan protocolos de comunicación digital, sin embargo para este laboratorio nos vamos a concentra en sensores que trabajan de forma nativa con las funciones de Digital Write y Read.

## Laboratorio de Relay

![Relay](https://cdn2.bigcommerce.com/server3700/64749/products/953/images/3093/1467712909__82097.1472166102.1280.1280.jpg?c=2)

Para utilizar un relay debemos realizar la siguiente conexión:

**Desconecte la Playground del cable USB**.

- Pin negativo en el relay, a GND en la Playground
- Pin positivo en el relay, a VOUT en la Playground
- Pin S (de "señal") a cualquier pin manipulable por DigitalWrite

Una vez realizadas las conexiones físicas, y pidiendo a un compañero que revise las conexiones, conecte de nuevo la Playground, y cargue el ejemplo anterior, modificando el PIN, de forma que sea el que usó para conectar el Relay.

2. Cuando se precione el botón B, se apaga el Relay

### Normalmente-conectado (NC), normalemente-abierto (NO)

Dependiendo del tipo de dispositivo que deseamos controlar con el relay, tomamos la fase positiva, y la conectamos a lo puertos del relay, ya sea en modo de NC o NO.

- NC: Si se apaga el Playground, el dispositivo conectado al Relay va a quedar en posición activo sin importar el código. Se puede utilizar para luces de emergencia.
- NO: Si se apaga el Playground, el dispositivo conectado al Relay va a quedar apagado. Útil para bombas de agua, llavines, calentadores, etc.

## Laboratorio de nivel de agua

![Sensor de nivel de agua](https://cdn2.bigcommerce.com/server3700/64749/products/1192/images/3916/Screen_Shot_2017-07-14_at_3.17.53_PM__38468.1500067089.1280.1280.png?c=2)

El leer nuestro sensor de nivel de agua también es una tarea muy sencilla porque lo único que necesitamos realizar es un DigitalRead, donde lo vamos a tratar como dispotivos estilo botón, switch o interruptor.

**Desconecte la Playground del cable USB**.

- Uno de los pines del sensor, a GND en la Playground
- El otro pin, a cualquier pin disponible por DigitalRead

Como ya hemos programado un poco, vamos a introducir una directiva llamada **If-then-else**, la cual evalúa una condición, en este caso el switch del nivel agua, y si está por encima del nivel, prende las luces de color rojo, y si la condición es contraria, ejecuta el Clear.
